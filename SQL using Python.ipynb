```python
from faker import Faker
import random
import mysql.connector

fake = Faker()

# connect to the MySQL database
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="",
  database="products_management"
)

# create a cursor object
mycursor = mydb.cursor()

# generate data for the product table
for i in range(1000):
    name = fake.word()
    category = random.choice(['Electronics', 'Clothing', 'Food', 'Beauty'])
    description = fake.text()
    price = round(random.uniform(1, 100), 2)
    stock_level = random.randint(0, 100)
    sql = "INSERT INTO product (name, category, description, price, stock_level) VALUES (%s, %s, %s, %s, %s)"
    val = (name, category, description, price, stock_level)
    mycursor.execute(sql, val)

# generate data for the supplier table
for i in range(1000):
    name = fake.company()
    address = fake.address()
    phone_number = fake.phone_number()
    email = fake.email()
    sql = "INSERT INTO supplier (name, address, phone_number, email) VALUES (%s, %s, %s, %s)"
    val = (name, address, phone_number, email)
    mycursor.execute(sql, val)

# generate data for the inventory table
for i in range(1000):
    product_id = random.randint(1, 1000)
    quantity = random.randint(0, 100)
    reorder_point = random.randint(0, 50)
    reorder_quantity = random.randint(1, 20)
    sql = "INSERT INTO inventory (product_id, quantity, reorder_point, reorder_quantity) VALUES (%s, %s, %s, %s)"
    val = (product_id, quantity, reorder_point, reorder_quantity)
    mycursor.execute(sql, val)

# generate data for the sale table
for i in range(1000):
    product_id = random.randint(1, 1000)
    quantity = random.randint(1, 10)
    price = round(random.uniform(1, 100), 2)
    date = fake.date_between(start_date='-1y', end_date='today')
    sql = "INSERT INTO sale (product_id, quantity, price, date) VALUES (%s, %s, %s, %s)"
    val = (product_id, quantity, price, date)
    mycursor.execute(sql, val)

# generate data for the customer table
for i in range(1000):
    name = fake.name()
    address = fake.address()
    phone_number = fake.phone_number()
    email = fake.email()
    sql = "INSERT INTO customer (name, address, phone_number, email) VALUES (%s, %s, %s, %s)"
    val = (name, address, phone_number, email)
    mycursor.execute(sql, val)

# commit the changes to the database
mydb.commit()

# close the cursor and database connection
mycursor.close()
mydb.close()
```

```python
from faker import Faker
import mysql.connector

# create a Faker object
fake = Faker('en_IN')

# create a MySQL connection
cnx = mysql.connector.connect(user='root', password='',
                              host='localhost', database='products_db')
cursor = cnx.cursor()

# generate data for the product table
for i in range(1000):
    name = fake.name()
    category = fake.random_element(elements=('electronics', 'clothing', 'home', 'beauty'))
    description = fake.text(max_nb_chars=200)
    price = fake.random_int(min=10, max=1000)
    stock_level = fake.random_int(min=1, max=100)
    query = "INSERT INTO product (name, category, description, price, stock_level) VALUES (%s, %s, %s, %s, %s)"
    values = (name, category, description, price, stock_level)
    cursor.execute(query, values)

# generate data for the supplier table
for i in range(1000):
    name = fake.name()
    address = fake.address()
    phone_number = fake.phone_number()
    email = fake.email()
    query = "INSERT INTO supplier (name, address, phone_number, email) VALUES (%s, %s, %s, %s)"
    values = (name, address, phone_number, email)
    cursor.execute(query, values)

# generate data for the inventory table
for i in range(1000):
    product_id = fake.random_int(min=1, max=1000)
    quantity_on_hand = fake.random_int(min=1, max=100)
    reorder_point = fake.random_int(min=1, max=50)
    reorder_quantity = fake.random_int(min=10, max=100)
    query = "INSERT INTO inventory (product_id, quantity_on_hand, reorder_point, reorder_quantity) VALUES (%s, %s, %s, %s)"
    values = (product_id, quantity_on_hand, reorder_point, reorder_quantity)
    cursor.execute(query, values)

# generate data for the sale table
for i in range(1000):
    product_id = fake.random_int(min=1, max=1000)
    quantity_sold = fake.random_int(min=1, max=10)
    price = fake.random_int(min=10, max=1000)
    date = fake.date_between(start_date='-1y', end_date='today')
    query = "INSERT INTO sale (product_id, quantity_sold, price, date) VALUES (%s, %s, %s, %s)"
    values = (product_id, quantity_sold, price, date)
    cursor.execute(query, values)

# generate data for the customer table
for i in range(1000):
    name = fake.name()
    address = fake.address()
    phone_number = fake.phone_number()
    email = fake.email()
    query = "INSERT INTO customer (name, address, phone_number, email) VALUES (%s, %s, %s, %s)"
    values = (name, address, phone_number, email)
    cursor.execute(query, values)

# commit the changes and close the connection
cnx.commit()
cursor.close()
cnx.close()
```

```python
import mysql.connector

class ProductsManager:
    #def __agh__
    def __init__(self, host, user, password, database):
        self.db = mysql.connector.connect(
            host=host,
            user=user,
            password=password,
            database=database
        )
        self.cursor = self.db.cursor()
        # self.__variable1__ = 12
        
    def add_product(self, name, category, description, price, stock_level):
        sql = "INSERT INTO product (name, category, description, price, stock_level) VALUES (%s, %s, %s, %s, %s)"
        values = (name, category, description, price, stock_level)
        self.cursor.execute(sql, values)
        self.db.commit()
        print(f"Added product {name} to the database.")
        
    def update_stock_level(self, product_id, new_stock_level):
        sql = "UPDATE inventory SET quantity_on_hand = %s WHERE product_id = %s"
        values = (new_stock_level, product_id)
        self.cursor.execute(sql, values)
        self.db.commit()
        print(f"Updated stock level of product with ID {product_id} to {new_stock_level}.")
        
    def add_supplier(self, name, address, phone_number, email):
        sql = "INSERT INTO supplier (name, address, phone_number, email) VALUES (%s, %s, %s, %s)"
        values = (name, address, phone_number, email)
        self.cursor.execute(sql, values)
        self.db.commit()
        print(f"Added supplier {name} to the database.")
        
    def view_sales_performance(self, product_id=None, category=None):
        if product_id is not None:
            sql = "SELECT * FROM sale WHERE product_id = %s"
            values = (product_id,)
        elif category is not None:
            sql = "SELECT * FROM sale JOIN product ON sale.product_id = product.id WHERE product.category = %s"
            values = (category,)
        else:
            print("Please provide either a product ID or a category.")
            return
        self.cursor.execute(sql, values)
        result = self.cursor.fetchall()
        if len(result) == 0:
            print("No sales data found.")
        else:
            print("Sales data:")
            for row in result:
                print(f"Product ID: {row[1]}, Quantity Sold: {row[2]}, Price: {row[3]}, Date: {row[4]}")

if __name__ == '__main__':
    pm = ProductsManager(host='localhost', user='root', password='', database='products_db')
    pm.add_product('Product A', 'Category A', 'Description A', 9.99, 100)
    pm.update_stock_level(1, 50)
    pm.add_supplier('Supplier A', 'Address A', '1234567890', 'suppliera@example.com')
    pm.view_sales_performance(product_id=1)
    pm.view_sales_performance(category='Category A')
```

```python
import mysql.connector

class ProductSales:
    
    def __init__(self, host, user, password, database):
        self.db = mysql.connector.connect(
            host=host,
            user=user,
            password=password,
            database=database
        )
        self.cursor = self.db.cursor()
        
    def get_sales_by_product(self, product_name):
        query = f"SELECT SUM(quantity), SUM(total_amount) FROM sales WHERE product_name = '{product_name}'"
        self.cursor.execute(query)
        result = self.cursor.fetchone()
        return f"Total Quantity Sold: {result[0]}, Total Sales Amount: {result[1]}"
    
    def get_sales_by_category(self, category):
        query = f"SELECT SUM(quantity), SUM(total_amount) FROM sales JOIN products ON sales.product_id = products.id WHERE category = '{category}'"
        self.cursor.execute(query)
        result = self.cursor.fetchone()
        return f"Total Quantity Sold: {result[0]}, Total Sales Amount: {result[1]}"
```
